```python
#Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Import the dataset.
# Assign it to a variable called crime.
crime = pd.read_csv("/content/US_Crime_Rates_1960_2014.csv")
```

```python
crime.head()
```

```python
# What is the type of the columns? integers
crime.info()
```

```python
# Convert the type of the column Year to datetime64
crime["Year"] = pd.to_datetime(crime["Year"])
```

```python
# Set the Year column as the index of the dataframe
crime.set_index("Year", inplace=True)
crime.head()
```

```python
#  Delete the Total column
crime = crime.drop(["Total"], axis=1)
```

```python
crime.head()
```

```python
# Group the year by decades and sum the values
# Pay attention to the Population column number, summing this column is a mistake
total = crime.groupby('Year').sum()
total.head()
```
